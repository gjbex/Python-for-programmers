```python
import matplotlib.pyplot as plt
import math
import random
import statistics
import typing
```

```python
Deck: typing.TypeAlias = typing.Sequence[int]
```

```python
def count_increasing_cards(deck: Deck) -> int:
    nr_cards = 1
    while nr_cards < len(deck) and deck[nr_cards] > deck[nr_cards - 1]:
        nr_cards += 1
    return min(len(deck), nr_cards + 1)
```

```python
def mean_cards(nr_shuffles: int =100_000, deck_size: int = 2) -> float:
    deck = list(range(deck_size))
    card_runs = []
    for _ in range(nr_shuffles):
        random.shuffle(deck)
        card_runs.append(count_increasing_cards(deck))
    return statistics.mean(card_runs)
```

```python
nr_shuffles = 1_000_000
max_deck_size = 30
sizes = list(range(1, max_deck_size))
runs = [mean_cards(nr_shuffles=nr_shuffles, deck_size=size) for size in sizes]
```

```python
plt.plot(sizes, runs)
plt.plot(sizes, [math.exp(1.0)]*len(sizes));
```
